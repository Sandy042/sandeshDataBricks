### COPY INTO

This particular command allows us to load the data from the file location,cloud storage into the delta table.

Query with copy into will have idempoteny means we will not get any duplicate data during the execution



Lets create a Volume : copyint

```python
%sql
CREATE VOLUME sandeshmsdatabricks.autoloader.copyintvol
```

- we are using the parquet file as a source file.

_why parquet file? because parquet file will be having the schema wrt the dataset and COPY INTO mainly depends on the schema of the dataset_

lets query the source file to see the column name and datatype so that we can create a table with correct schema

```python
%sql
select * from parquet.`/Volumes/sandeshmsdatabricks/autoloader/copyintvol/part-00000-tid-1800520145580676501-2fb00642-a896-4599-9957-2c7260f04aa7-132-1.c000.snappy.parquet`
```

### Create table

As COPY INTO works well on table we are creating the table wrt the dataset/source file with valid schema.

```python
%sql
CREATE TABLE sandeshmsdatabricks.autoloader.tborder(
  order_id INT,
  customer_id INT,
  order_date DATE,
  order_status string,
  order_amount DOUBLE
)
```

### COPY INTO Query

Below query acts as incremental loader from source to the delta table, here it will not copy the same data multiple times, it make sure to keep idempotency

```python
%sql
COPY INTO sandeshmsdatabricks.autoloader.tborder
FROM '/Volumes/sandeshmsdatabricks/autoloader/copyintvol'
FILEFORMAT = PARQUET
FORMAT_OPTIONS ('mergeSchema' = 'true')
COPY_OPTIONS ('mergeSchema' = 'true')
```

```python
%sql
SELECT * FROM sandeshmsdatabricks.autoloader.TBORDER
```
