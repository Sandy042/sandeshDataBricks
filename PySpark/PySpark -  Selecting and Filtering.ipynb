```python
from pyspark.sql.functions import *
df=spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load('/Volumes/sandeshmsdatabricks/sourcefiles/sourcevolume/weather/Brazil_weather_data.csv')
```

### SELECTING and FILTERING

**select()** chooses the specific column.
- _Syntax: df.select('col1','col2','col(col3).alias("newnameforcol3"))_

**filter()/where()** removes rows based on condition

- _Syntax1: df.filter(col("col1")>10)_
- _Syntax2: df.filter("col1 >'10' and col2='X')_
- 
Foundation of all DataFrame transformationsâ€”reduces data movement, optimizes Spark execution.

**Dataset** used : Brazil_weather_data.csv

### Question 1
Select Country, Date, Temp_Mean, Windgusts_Max. Filter for Temp_Mean > 25 AND Windgusts_Max > 10.

```python
df.select("Country","Date","Temp_Mean","Windgusts_Max")\
.filter("Temp_Mean>25 and Windgusts_Max>10")\
.show()
```

### Question 2
Select Date, _Precipitation_Sum_. Filter Precipitation_Sum > 5 OR Windgusts_Max > 20. Limit 10 rows.

```python
df.select("Date","Precipitation_Sum")\
  .filter("Precipitation_Sum>5 or Windgusts_Max > 20")\
  .limit(10)\
  .show()
```

### Question 3

Select ALL columns containing "Temp" OR "Wind" using regex. Filter Temp_Mean > 20 AND Country = "Brazil". Show 5 rows


```python
df.select([col(x) for x in df.columns if any(pattern in x for pattern in ["Temp", "Wind"])])\
.filter("Temp_Mean > 20 AND Country = 'Brazil'")\
.limit(5)\
.show()
```
